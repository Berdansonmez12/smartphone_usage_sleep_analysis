```python
import pandas as pd

df = pd.read_csv("DSA210 - Sheet.csv")
df.head()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df["screen after 23"], df["sleep duration"])
plt.xlabel("Screen After 23")
plt.ylabel("Sleep Duration")
plt.title("Late Night Usage vs Sleep Duration")
plt.show()
```

```python
df.corr(numeric_only=True)
```

## Insights

The analysis shows a weak negative relationship between late-night screen usage (after 23:00) and sleep duration. This suggests that increased late-night phone usage may reduce total sleep time.

Additionally, screen usage after 22:00 and 23:00 are highly correlated (0.86), indicating that users who use their phones late tend to continue using them into the night.

Interestingly, total screen time shows a moderate positive correlation with sleep duration, which may indicate that daytime usage does not negatively impact sleep as much as late-night usage.

Notifications and unlocks are also moderately correlated (0.51), suggesting that increased notifications lead to more frequent phone checking behavior.
